```python
from pytweezer.experiment import Experiment, DummyCamera, DummyAnalyser, DummyExperimentRunner, dummy_analysis
```

```python
experiment_runner = DummyExperimentRunner()
camera = DummyCamera()
analyser = DummyAnalyser()
experiment = Experiment(
    experiment_runner=experiment_runner,
    script="dummy_script.mmp",
    camera=camera,
    analyser=analyser,
    n_images=2,
    analysis_funcs=[dummy_analysis],
)
results = experiment.run(
    scan_param="param1", scan_vals=[1, 2, 3], n_repeats=2, save=False
)
```

```python
import h5py
with h5py.File("experiment_result_6.h5", "r") as f:
    print("Experiment ID:", f.attrs['rid'])
    print("Scan param:", f.attrs['scan_param'])
    print("Scan vals:", f['scan_vals'][:])
    # print("Images:", f['images'][:])
    for key in f.keys():
        if key.startswith('analysis_result_'):
            print(f"{key}:", f[key][:])
    print("Zips:", f['zips'][:])
```
