# Test SLM Control

```python
import pytweezer.phasemask as pm
import pytweezer.communication as com
import numpy as np
import cupy as cp
import matplotlib.pyplot as plt
```

```python
PM = pm.OptimisationBasedPhasemaskGeneratorGPU(
                 wavelength_um=0.852,
                 focal_length_mm=17.3,
                 slm_pitch_um=17,
                 slm_res=(1024,1024),
                 input_beam_waist_mm=16,
                 fresnel_f_mm=1070,
                 blaze_dx_dy_um=(40.0, 0.0))

SLM = com.SLMClient()

pm_init = PM.superimpose([PM.fresnel, PM.blaze])
pm_init_uint8 = PM.transform_phase_8bit(pm_init).get()
SLM.update_mask(pm_init_uint8)
```

```python
W = np.ones((8, 8))
spacing_um = 5

target = PM.generate_weighted_array(W, spacing_um, init_phase_randomness=1.0)
pm_slm, terms, _ = PM.superposition_optimization(target, max_iter=500, damping=0.5, verbose=True)
```

```python
PM = pm.OptimisationBasedPhasemaskGeneratorGPU(
                 wavelength_um=0.852,
                 focal_length_mm=17.33,
                 slm_pitch_um=17,
                 slm_res=(1024,1024),
                 input_beam_waist_mm=16,
                 fresnel_f_mm=1040,
                 blaze_dx_dy_um=(40.0, 0.0))

pm_init = PM.superimpose([pm_slm, PM.fresnel, PM.blaze])
pm_init_uint8 = PM.transform_phase_8bit(pm_init).get()
SLM.update_mask(pm_init_uint8)
```
