```python
import zmq
import numpy as np
import time
import json

ZMQ_PORT = 5555
IMG_SHAPE = (512, 512, 3)
GROUP_SIZE = 3

context = zmq.Context()
publisher = context.socket(zmq.PUB)
publisher.bind(f"tcp://*:{ZMQ_PORT}")

time.sleep(1)  # let viewer connect


```

```python
from multiprocessing import Process
from pytweezer.image_viewer import run_viewer  # your viewer function

# Start the viewer in a separate process
viewer_process = Process(target=lambda: run_viewer(group_size=2))
viewer_process.start()
```

```python
for i in range(30):
    img = np.random.randint(0, 255, IMG_SHAPE, dtype=np.uint8)
        # metadata header
    header = {
        "shape": img.shape,
        "dtype": str(img.dtype),
        "index": i,
        "description": f"Batch {i}"
    }
    header_bytes = json.dumps(header).encode("utf-8")
    publisher.send_multipart([header_bytes, img.tobytes()])
    time.sleep(0.2)
#
```
