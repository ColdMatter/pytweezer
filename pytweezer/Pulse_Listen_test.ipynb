```python
import sys

sys.path.insert(0, "pytweezer")

from pytweezer.arduino import ArduinoPulser

arduino = ArduinoPulser("COM4", baudrate=9600, timeout=1)

startup_message = arduino.connect()
print(f" Arduino: '{startup_message}'")
```

```python
import time

number_of_frames = 10


def load_images_to_slm():
    print(f"{number_of_frames} images loaded to SLM memory.")


load_images_to_slm()

confirmation = arduino.send_pulses(number_of_frames)
print(confirmation)

time.sleep(1)
```

```python
shutdown_message = arduino.close()
print(f"Arduino: '{shutdown_message}'")
print("Port closed successfully.")
```
